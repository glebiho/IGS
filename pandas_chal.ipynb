```python
#Recodage de sexe (0 pour masculin et 1 pour féminin)
df["sexe"] = df["sexe"].replace(["masculin","feminin"], [0,1])
df
```

```python
#supprimer les variable depression, vaisseau et pic du dataframe
sup = df.drop(["depression","vaisseau","pic"], axis = 1)
sup
```

```python
#Representation graphique de la variable coeur
plt.figure()
df['coeur'].value_counts().plot.pie()
plt.show()
```

```python
#Representation graphique de toutes les variables qualitatives
list_qual=['sucre', 'electro', 'angine', 'coeur']
plt.figure(figsize=(17,7))
i=1
for element in list_qual:
    plt.subplot(1,7,i)
    df[element].value_counts().plot(kind='pie')
    i+=1
    plt.title(element)
```

```python
#Representation graphique de toutes les variables quantitative
list_quan=['age', 'sexe', 'pression', 'cholester', 'taux_max']
plt.figure(figsize=(40,10))
i=1
for a in list_quan:
    plt.subplot(2,3,i)
    df[a].value_counts().plot(kind='hist',color='gray')
    i+=1
    plt.title(a)
```
